```python

#converting data type
df['Aggregate rating']=df['Aggregate rating'].astype('int64')
     

# Display basic statistics of the target variable
print("Basic Statistics of Aggregate Rating:")
print(df['Aggregate rating'].describe())

# Plot the distribution of the target variable
plt.figure(figsize=(8, 6))
df['Aggregate rating'].hist(bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Aggregate Rating')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.show()

# Check for class imbalances
class_counts = df['Aggregate rating'].value_counts()
print("Class Counts:")
print(class_counts)

# Plot the class distribution
plt.figure(figsize=(8, 6))
class_counts.plot(kind='bar', color='lightcoral')
plt.title('Class Distribution of Aggregate Rating')
plt.xlabel('Aggregate Rating')
plt.ylabel('Count')
plt.show()
```
